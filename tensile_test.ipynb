```python
import os
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

Extract Tensile stress and strain from .csvs and plot

```python
def stress_v_strain(export_dir):

  all_runs = []
  strain = [] # strain for all runs
  stress = [] # stress for all runs

  for run in os.listdir(export_dir):
    if run.endswith('_2.csv'):
      run_path = os.path.join(export_dir,run)
      df = pd.read_csv(run_path)
      df = df.drop(index = 0)
      strain_r = df['Tensile strain (Displacement)'].astype(float) # tensile strain (N)
      stress_r = df['Tensile stress'].astype(float) # tensile stress (MPa)
      strain.append(strain_r)
      stress.append(stress_r)
      all_runs.append(run)
      #print(df.head())

  sns.set_style('whitegrid')
  sns.set_palette('Spectral',9)
  plt.figure(figsize=(12,6))

  for x,y,r in zip(strain,stress,all_runs):
    run = ''.join(r.split('-')[3:5]).rstrip('_2.csv') #remove extra text to give labels for each run
    plt.plot(x,y, label = run)

  plt.xlabel('Tensile Strain (%)', fontsize = 18)
  plt.ylabel('Tensile Stress (MPa)', fontsize = 18)
  plt.xticks(fontsize = 16)
  plt.yticks(fontsize = 16)
  plt.xlim(0, 21)
  plt.ylim(0, 145)
  plt.legend(fontsize = 16)
  plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
  #plt.savefig('figures/'+export_name+'.png',dpi = 300) #define export_name to be something clever for each plot
  plt.show()
```

Test plotting

```python
export_dir = '/content/drive/MyDrive/Tensile Testing 2 23/Hayley 2023-02-22-DFS10-12.is_tens_Exports'

stress_v_strain(export_dir)
```

```python
tensile_dir = '/content/drive/MyDrive/Tensile Testing 2 23'

all_exports = []

for export in os.listdir(tensile_dir):
  all_exports.append(os.path.join(tensile_dir,export))

for export in all_exports:
  stress_v_strain(export)

```
